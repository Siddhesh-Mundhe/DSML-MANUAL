```python
#import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv("/content/heart.csv")
file.head()
```

```python
# Sactterplot
x= file['chol']
y = file['thalach']
plt.scatter(x,y, color='r', s=10)
plt.xlabel("chol")
plt.ylabel("thalach")
plt.title("Scatter Plot")
sns.set_style('whitegrid')
plt.show()
```

```python
# Histogram
x = file['chol']
plt.hist(x, bins=50)
plt.xlabel("chol")
plt.title("Histogram")
plt.show()
```

```python
sns.boxplot(x = "thalach", data = file, width=0.8)
plt.xlabel('thalach')
plt.title('Box Plot on thalach')
```

```python
# bar graph
x = file["thalach"]
y = file["chol"]
plt.bar(x,y, color='r', width=0.5)
plt.xlabel("chol")
plt.ylabel("thalach")
plt.title("Bar Graph")
plt.show()
```

```python
x = [file['thalach'], file['restecg']]
plt.imshow(x ,interpolation="nearest", cmap="hot")
plt.title("Heatmap")
plt.colorbar()
plt.show()
```
